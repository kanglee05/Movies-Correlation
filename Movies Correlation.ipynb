```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data

df = pd.read_csv('/Users/kanglee/Downloads/movies.csv')
```

```python
# Quick look at the data

df.head()
```

```python
# Check for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Drop missing data rows

df = df.dropna()
```

```python
# Check rows were correctly dropped

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data types for columns

df.dtypes
```

```python
# Change data type

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Create correct year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
# Sort by gross descending

df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
# Correlations

# Select only numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=['number'])

# Compute the Pearson correlation on the numeric columns
correlation_matrix = numeric_df.corr(method='pearson')

# Display the correlation matrix
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
# High correlation between movie budget and gross earnings
```

```python
# Scatter plot with score vs gross

plt.scatter(x=df['score'], y=df['gross'])

plt.title('Score vs Gross earnings')

plt.xlabel('Score for Movie')

plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Plot score vs gross using seaborn

sns.regplot(x='score', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```
